```python
class LenearRegression:
    
    def __init__(self, lr=.01, max_steps=2000):
        self.lr = lr
        self.max_step = max_steps
        
    def fit(self, x, y):
        self.cofficient = np.random.random() ## cofficent here is m
        self.intercept = np.random.random()## intercept here is c
        
        for i in range(self.max_step):
            self.gradient_decend(x, y)
            
    def gradient_decend(self, x, y):
        delta_cofficient, delta_intercept = self.gradient(x, y)
        #print(delta_cofficient)
        self.cofficient += delta_cofficient
        self.intercept += delta_intercept
        
    def gradient(self, x, y):
        yp = self.predict(x)
        d_e2m = np.mean((yp - y)*x)
        d_e2c = np.mean((yp - y))
        
        return -d_e2m * self.lr, -d_e2c * self.lr
    
    def predict(self, x):
        return self.intercept + self.cofficient * x
```

```python
model = LenearRegression()
```

```python
model.fit(x, y)
```
