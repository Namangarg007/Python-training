```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = Image.open("../datasets/monastery.jpg")

small = img.resize(size=(400, 300))

small

img = np.array(small).mean(axis=2)

plt.imshow(img, cmap="gray")
```

```python
result = np.zeros([img.shape[0]//2, img.shape[1]//2 ])
```

```python
for row in range(result.shape[0]):
    for col in range(result.shape[1]):
        cut = img[2*row:2*row+2, 2*col:2*col+2]
        result[row, col] = np.max(cut)

plt.imshow(result, cmap="gray")
```

```python
for row in range(result.shape[0]):
    for col in range(result.shape[1]):
        cut = img[2*row:2*row+2, 2*col:2*col+2]
        result[row, col] = np.mean(cut)

plt.imshow(result, cmap="gray")
```
