```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = Image.open("../datasets/monastery.jpg")

small = img.resize(size=(400, 300))

small

img = np.array(small).mean(axis=2)

plt.imshow(img, cmap="gray")
```

```python
padded_image = np.pad(img, pad_width=1)
```

```python
img.shape
```

```python
padded_image.shape
```

```python
plt.imshow(padded_image, cmap="gray")
```

# Horizontal Edge Detection

```python
fil = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])

result = np.zeros_like(img)

for row in range(result.shape[0]):
    for col in range(result.shape[1]):
        cut = padded_image[row:row+3, col:col+3]
        result[row, col] = abs((cut * fil).mean())

plt.imshow(result, cmap="gray")
```

# Vertical Edge Detection

```python
fil = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])

result = np.zeros_like(img)

for row in range(result.shape[0]):
    for col in range(result.shape[1]):
        cut = padded_image[row:row+3, col:col+3]
        result[row, col] = abs((cut * fil).mean())

plt.imshow(result, cmap="gray")
```
