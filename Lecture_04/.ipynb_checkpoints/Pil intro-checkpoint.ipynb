```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros((100, 100, 3), dtype=np.uint8) # uint stands for unsigned int
```

```python
plt.imshow(img)
```

```python
img[:30, : , 0] = 255
```

```python
plt.imshow(img)
```

```python
img[30:70, :, 1] = 255
```

```python
plt.imshow(img)
```

```python
plt.imshow(img)
```
