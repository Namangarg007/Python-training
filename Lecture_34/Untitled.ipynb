```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
```

```python
img = Image.open("Doctor.jpg")
```

```python
np_img = np.array(img)
```

```python
plt.imshow(np_img)
```

```python
np_img.dtype
```

```python
np_img.shape
```

```python
X = np_img.reshape(-1, 3)
```

```python
X.shape
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(30)
```

```python
model.fit(X)
```

```python
pallets = model.cluster_centers_.astype(np.uint8)
```

```python
pallets
```

```python
plt.imshow(pallets.reshape(1, -1, 3))
```

```python
output = np.array([pallets[clas] for clas in model.labels_])
```

```python
out_img = output.reshape(np_img.shape)
```

```python
plt.imshow(out_img)
```
