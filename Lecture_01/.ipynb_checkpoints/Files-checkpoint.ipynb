```python
f = open("hello.txt", "wt")#wt means write text and open is used to open a specific file and if file doesnot exit then it will only be created only if it is in write mode
```

```python
f.write("hello World") #will wite specifed text in the file reference in f
```

```python
f.close()#to close the file and save the file with entered data
```

```python
import requests  #will help fetch some thing or some url
```

```python
url = r"https://www.google.com/" #r stands for raw string it will print the whole string if there are special symbols in the string
```

```python
result = requests.get(url)
```

```python
f = open("google.html", "wt")

f.write(result.text)

f.close()
```

```python
url = r"https://i.ytimg.com/vi/v_cpPMjE0vU/maxresdefault.jpg"
result = requests.get(url)

f = open("panda.jpg", "wb")#wb stands for write baiary

f.write(result.content)# to write content .content extension is used

f.close()
```
