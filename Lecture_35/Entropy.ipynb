```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bullets = 10

entropy = []

for fake in range(0, 11):
    
    real = bullets - fake
    
    data = ["r"] * real + ["f"] * fake
    items, count = np.unique(data, return_counts=True)
    
    p = count/len(data)
    
    entropy.append(np.sum(-p * np.log(p)))
```

```python
plt.plot(entropy)
```
