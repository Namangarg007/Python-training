```python
class cluster:
    
    def __init__(self, center, name):
        
        self.center = center
        self.name = name
        self.point = []
        
    def distance(self, points):
        
        return np.sqrt(np.sum((self.center - points) ** 2))
    
    def relocate(self):
        
        self.center = np.mean(self.point, axis=0)
        self.point.clear()
```
