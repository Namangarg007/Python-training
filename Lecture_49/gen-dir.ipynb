```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from tensorflow.keras.preprocessing.image import load_img

import numpy as np
import matplotlib.pyplot as plt
```

```python
generator = ImageDataGenerator(rotation_range=10, horizontal_flip=True, shear_range=1.1)
```

```python
gen = generator.flow_from_directory("./data/")
```

```python
gen.class_indices
```

```python
x, y = gen.next()
```

```python
for gen_img, index in zip(x, y):
        plt.figure()
        plt.title(str(index))
        plt.imshow(gen_img.astype(int))
        plt.show()
```
